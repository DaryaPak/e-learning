## Сколько студентов успешно сдали только один курс?

Объединяю таблицы, чтобы отфильтровать строки только с экзаменами:

```python
full_assessments = student_assessment.merge(assessments)
full_assessments
```

На один курс могло приходиться больше одного экзамена, поэтому считать по кол-ву сданных экзаменов - не лучшая идея:

```python
full_assessments \
.query('assessment_type == "Exam"') \
.groupby('code_module', as_index = False) \
.agg({'id_assessment' : 'nunique'})
```

Создаю таблицу с данными только об экзаменах:

```python
exams = full_assessments.query('assessment_type == "Exam"')
```

Добавляю столбец, который отражает, сдал ли студент экзамен:

```python
exams = exams.assign(success = exams.score >= 40)
exams.head()
```

Сколько студентов успешно завершили только один курс:

```python
exams \
.query('success == True') \
.groupby('id_student', as_index = False) \
.agg({'code_module':'count'}) \
.query('code_module == 1') \
.shape[0]
```

## Выяви самый сложный и самый простой экзамен: найди курсы и экзамены в рамках курса, которые обладают самой низкой и самой высокой завершенностью

***Завершенность = количество успешных экзаменов / количество всех попыток сдать экзамен***

Количество успешных попыток в разбивке по курсу и экзамену:

```python
success = exams \
.query('success == True') \
.groupby(['code_module', 'id_assessment'], as_index = False) \
.agg({'id_student' : 'count'}) \
.rename(columns={'id_student' : 'success_count'})
success
```

Количество всех попыток:

```python
total_attempts = exams \
.groupby(['code_module', 'id_assessment'], as_index = False) \
.agg({'id_student' : 'count'}) \
.rename(columns={'id_student' : 'total_attempts'})
total_attempts
```

Объединяю таблицы:

```python
df_exams = success.merge(total_attempts, on=['code_module', 'id_assessment'])
df_exams
```

Создаю финальную таблицу с новым столбцом, содержащим процент завершенности:

```python
complete_ex_df = df_exams.assign(complete = round(df_exams.success_count / df_exams.total_attempts * 100, 2)) \
.groupby(['code_module', 'id_assessment'], as_index = False) \
.agg({'complete' : 'mean'})
complete_ex_df
```

Самый сложный экзамен:

```python
complete_ex_df[complete_ex_df['complete'] == complete_ex_df['complete'].min()]
```

Самый легкий экзамен:

```python
complete_ex_df[complete_ex_df['complete'] == complete_ex_df['complete'].max()]
```

Завершаемость каждого курса в целом:

```python
complete_ex_df \
.groupby('code_module', as_index = False) \
.agg({'complete': 'mean'})
```

Ответы по заданию 2:  
Самый сложный экзамен: **25340**  
Самый простой экзамен: **25361**  
Курс с наиболее высокой завершенностью: **DDD**  
Курс с наименее высокой завершенностью: **CCC**

## Выяви самые популярные предметы (ТОП-3) по количеству регистраций на них, а также предметы с самым большим оттоком (ТОП-3)

Считаю количество регистраций на каждый курс(предмет), сортирую и выбираю ТОП-3:

```python
student_registration \
.value_counts('code_module') \
.head(3) \
.to_frame(name='registrations') \
.reset_index() 
```

***Отток = количество студентов, отменивших регистрацию / количество студентов, которые регистрировались***

Сколько всего людей зарегестрировалось на какой-либо курс:

```python
total = student_registration \
.groupby('code_module', as_index = False) \
.agg({'date_registration' : 'count'}) \
.rename(columns = {'date_registration': 'total'})
total
```

Сколько отменили регистрацию (не берем в расчет значения NaN, т.к. эти студенты окончили курс):

```python
unreg = student_registration \
.query('date_unregistration != "NaN"') \
.groupby('code_module', as_index = False) \
.agg({'date_unregistration' : 'count'}) \
.rename(columns = {'date_unregistration': 'unreg'})
unreg
```

Объединяю таблицы:

```python
churn_df = total.merge(unreg)
```

Создаю новый столбец с показателем оттока студентов, сортирую по самому высокому оттоку:

```python
churn_df \
.assign(churn = round(churn_df.unreg / churn_df.total * 100, 2)) \
.sort_values('churn', ascending = False) \
.reset_index(drop=True) \
.head(3)
```

Ответы по заданию 4:  
ТОП-3 предметов по количеству регистраций: **BBB, FFF, DDD**  
ТОП-3 предметов с самым большим оттоком: **CCC, DDD, FFF**

## Используя python, построй адаптированные RFM-кластеры студентов, чтобы качественно оценить свою аудиторию

Среднее время сдачи одного экзамена:

```python
r = exams \
.groupby(['id_student'], as_index = False) \
.agg({'date_submitted' : 'mean'}) \
.rename(columns = {'date_submitted' : 'Recency'})
r
```

Количество успешных попыток сдать экзамен для каждого студента:

```python
student_success = full_last_exams \
.groupby(['id_student'], as_index = False) \
.agg({'id_assessment' : 'count'}) \
.rename(columns={'id_assessment' : 'success_count'})
student_success
```

Всего попыток для каждого студента:

```python
student_total_attempts = exams \
.groupby(['id_student'], as_index = False) \
.agg({'id_assessment' : 'count'}) \
.rename(columns={'id_assessment' : 'total_attempts'})
student_total_attempts
```

Объединяю таблицы:

```python
f = student_success.merge(student_total_attempts, how='outer', on='id_student')
f = f.fillna(0)
f
```

Рассчитываю показатель завершенности курсов каждым студентом:

```python
f = f.assign(Frequency = round(f.success_count / f.total_attempts * 100, 2)) \
.drop(['success_count', 'total_attempts'], axis=1)
f
```

Среднее количество баллов, получаемое за экзамен:

```python
m = exams \
.groupby('id_student', as_index = False) \
.agg({'score' : 'mean'}) \
.rename(columns = {'score' : 'Monetary'})
m
```

Объединяю все таблицы для расчета метрик:

```python
rf = r.merge(f)
rfm = rf.merge(m)
rfm
```

Границы для интерпретации кластеров:

```python
quintiles = rfm[['Recency', 'Monetary']].quantile([.2, .4, .6, .8]).to_dict()
quintiles
```

```python
rfm.Frequency.unique()
```

Есть три вида завершенности курсов (сдал с первого раза, со второго, не сдал).  
Метрика, подобранная для среднего времени сдачи и среднего балла, не подойдет для завершенности.

Метрика для оценки среднего срока сдачи экзамена: чем быстрее студент сдал экзамен, тем лучше, и тем выше оценка:

```python
def r_score(x):
    if x <= quintiles['Recency'][.2]:
        return 5
    elif x <= quintiles['Recency'][.4]:
        return 4
    elif x <= quintiles['Recency'][.6]:
        return 3
    elif x <= quintiles['Recency'][.8]:
        return 2
    else:
        return 1
```

Метрика для оценки завершенности студентами курсов: если сдал с первого раза, оценка выше:

```python
def f_score(x):
    if x == 100:
        return 2
    if x == 50:
        return 1
    else:
        return 0
```

Метрика для оценки среднего балла, получаемого за экзамен: чем выше средний балл, тем выше оценка:

```python
def m_score(x):
    if x <= quintiles['Monetary'][.2]:
        return 1
    elif x <= quintiles['Monetary'][.4]:
        return 2
    elif x <= quintiles['Monetary'][.6]:
        return 3
    elif x <= quintiles['Monetary'][.8]:
        return 4
    else:
        return 5
```

Применяю функции к таблице, рассчитывая метрики для каждого студента:

```python
rfm['R'] = rfm['Recency'].apply(lambda x: r_score(x))
rfm['F'] = rfm['Frequency'].apply(lambda x: f_score(x))
rfm['M'] = rfm['Monetary'].apply(lambda x: m_score(x))
rfm
```

Кластеры построены, оцениваю самые частые значения каждого из них:

```python
r_analysis = rfm \
.groupby(['R'], as_index = False) \
.agg({'id_student': 'count'}) \
.rename(columns={'id_student': 'cnt'}) \
.sort_values('cnt', ascending = False)
r_analysis.assign(percent = round(r_analysis.cnt / 4633 * 100, 2))
```

```python
f_analysis = rfm \
.groupby(['F'], as_index = False) \
.agg({'id_student': 'count'}) \
.rename(columns={'id_student': 'cnt'}) \
.sort_values('cnt', ascending = False)
f_analysis.assign(percent = round(f_analysis.cnt / 4633 * 100, 2))
```

```python
m_analysis = rfm \
.groupby(['M'], as_index = False) \
.agg({'id_student': 'count'}) \
.rename(columns={'id_student': 'cnt'}) \
.sort_values('cnt', ascending = False)
m_analysis.assign(percent = round(m_analysis.cnt / 4633 * 100, 2))
```
